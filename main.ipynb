```python
import pandas as pd
import csv
import numpy as np

data = pd.read_csv("sorted_constraints.csv") #name of input file
No_of_working_days = 6
No_of_lectures_per_day = 4
# print(data)
```

```python
#Extracting batch names from dataset

batches_df = data.drop('Teacher', axis = 'columns')
batches = batches_df.head(0)
# print(batches)
```

```python
class TimeTable:
    
    def __init__(self, data, batches, no_of_days, no_of_lectures):
        self.data = data
        self.batches = batches
        self.no_of_days = no_of_days
        self.no_of_lectures = no_of_lectures
        self.Teacher_list = self.createTeachers()
        self.Batch_list = self.createBatches()
    
    # defining an inner teacher class
    class Teacher:
    
        def __init__(self,teacher_name, no_of_days, no_of_lectures):
            self.name = teacher_name
            self.no_of_days = no_of_days
            self.no_of_lectures = no_of_lectures

        # generating a csv file for teacher's time table
        def createfile(self):
            daytt =[]
            days = []
            if self.no_of_days == 6:
                daytt = [{'Mon':'' , 'Tue':'' , 'Wed':'' , 'Thur':'' , 'Fri':'' , 'Sat':'' }]
                days = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat']
            elif self.no_of_days == 5:
                daytt = [{'Mon':'' , 'Tue':'' , 'Wed':'' , 'Thur':'' , 'Fri':''}]
                days = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri']
            elif self.no_of_days == 4:
                daytt = [{'Mon':'' , 'Tue':'' , 'Wed':'' , 'Thur':''}]
                days = ['Mon', 'Tue', 'Wed', 'Thur']
            elif self.no_of_days == 3:
                daytt = [{'Mon':'' , 'Tue':'' , 'Wed':''}]
                days = ['Mon', 'Tue', 'Wed']
            elif self.no_of_days == 2:
                daytt = [{'Mon':'' , 'Tue':''}]
                days = ['Mon', 'Tue']
            else:
                daytt = [{'Mon':''}]
                days = ['Mon']
                
            myTT = []
            
            for i in range(self.no_of_lectures):
                myTT+= daytt

            file_name = self.name + ".csv"

            with open(file_name, 'w') as csvfile:

                writer = csv.DictWriter(csvfile, fieldnames= days)
                writer.writeheader()
                writer.writerows(myTT)

        # check if the teacher is available at the given time slot
        def isAvai(self,slot, day):
            tt = pd.read_csv(self.name + ".csv")
            batch = tt.iat[slot, day]
            if pd.isna(batch):
                return True
            return False

        def getVal(self,slot, day):
            tt = pd.read_csv(self.name + ".csv")
            batch = tt.iat[slot, day]
            return batch

        # update the batch of the given time slot
        def updateSlot(self, slot, day, batch):
            tt = pd.read_csv( self.name + '.csv')
            tt.iat[slot, day] = batch
            days = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat']
            file_name = self.name + ".csv"

            tt.to_csv(file_name, index = False)
#             print(tt)
            return True

        def showTT(self):
            tt = pd.read_csv( self.name + '.csv')
            print(tt)
            
        def getTT(self):
            tt = pd.read_csv(self.name + '.csv')
            return tt
            
    #defining an inner batch class
    class Batch:

        def __init__(self, batch_name, no_of_days,no_of_lectures):
            self.name = batch_name
            self.no_of_days = no_of_days
            self.no_of_lectures = no_of_lectures
            self.status = False

        # generating a csv file for teacher's time table
        def createfile(self):
            daytt =[]
            days = []
            if self.no_of_days == 6:
                daytt = [{'Mon':'' , 'Tue':'' , 'Wed':'' , 'Thur':'' , 'Fri':'' , 'Sat':'' }]
                days = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat']
            elif self.no_of_days == 5:
                daytt = [{'Mon':'' , 'Tue':'' , 'Wed':'' , 'Thur':'' , 'Fri':''}]
                days = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri']
            elif self.no_of_days == 4:
                daytt = [{'Mon':'' , 'Tue':'' , 'Wed':'' , 'Thur':''}]
                days = ['Mon', 'Tue', 'Wed', 'Thur']
            elif self.no_of_days == 3:
                daytt = [{'Mon':'' , 'Tue':'' , 'Wed':''}]
                days = ['Mon', 'Tue', 'Wed']
            elif self.no_of_days == 2:
                daytt = [{'Mon':'' , 'Tue':''}]
                days = ['Mon', 'Tue']
            else:
                daytt = [{'Mon':''}]
                days = ['Mon']
                
            myTT = []
            
            for i in range(self.no_of_lectures):
                myTT+= daytt

            file_name = self.name + ".csv"

            with open(file_name, 'w') as csvfile:

                writer = csv.DictWriter(csvfile, fieldnames= days)
                writer.writeheader()
                writer.writerows(myTT)

        # check if the batch is free at the given time slot
        def isAvai(self,slot, day):
            tt = pd.read_csv(self.name + ".csv")
            lecture = tt.iat[slot, day]
            if pd.isna(lecture):
                return True
            return False

        # update the teacher of the given time slot
        def updateSlot(self, slot, day, teacher):
            tt = pd.read_csv( self.name + '.csv')
            tt.iat[slot, day] = teacher
            days = ['Mon', 'Tue', 'Wed', 'Thur', 'Fri', 'Sat']
            file_name = self.name + ".csv"

            tt.to_csv(file_name, index = False)
#             print(tt)
            return True

        # to get the status of the batch if its time table is done or not
        def getStatus(self):
            self.updateStatus()
#             if self.status == False:
#                 print(self.name)
            return self.status

        # update the status if batch time table is done(true) or not(false)
        def updateStatus(self):
            batch_tt = self.getTT()
            empty_cells = batch_tt.isnull().sum().sum()  # Counts the number of empty cells

            if empty_cells > 0:
                self.status = False
            else:
                self.status = True
        
        def showTT(self):
            tt = pd.read_csv( self.name + '.csv')
            print(tt)
            
        def getTT(self):
            tt = pd.read_csv(self.name + '.csv')
            return tt
        
    #Implementing other functions for time table generation
        
    #Instantiating objects
    def createTeachers(self):
        teacher_list = [self.Teacher(name, self.no_of_days, self.no_of_lectures) for name in self.data.Teacher]
        return (teacher_list)

    def createBatches(self):
        batch_list = [self.Batch(name, self.no_of_days, self.no_of_lectures) for name in self.batches]
        return batch_list
    
    def createAllFiles(self):
        for teacher in self.Teacher_list:
            teacher.createfile()
        for batch in self.Batch_list:
            batch.createfile()

    # checking if any batch contains an unalloted lecture
    def checkCompleted(self):
        df = self.data.drop(self.data.columns[0], axis=1, inplace=False)
        any_non_zero = (df != 0).any().any()
        return not any_non_zero

    #allocating lectures
    def allocateLectures(self):
#         self.allocateClassTeacher()
       # Iterate over each cell in the DataFrame
        for row_index, row in self.data.iterrows():
            col_index = 0
            teacher_name = ""
            lectures = 0

            for column_name, value in row.items():
#                 print(f"Cell ({row_index}, {column_name}, {col_index}): {value}")
                if col_index == 0:
                    teacher_name = value
                    col_index+=1
                    continue
                    
                teacher_index = row_index
                batch_index = col_index -1
                batch_name = column_name
                temp_value = value
                
#                 alloting 1st lecture to class teacher
                if value>=self.no_of_days and self.Batch_list[batch_index].isAvai(0,0) and self.Teacher_list[teacher_index].isAvai(0,0):
#                     print("I'm alloting class teacher")
                    for i in range(self.no_of_days):
                        self.Teacher_list[teacher_index].updateSlot(0,i,batch_name)
                        self.Batch_list[batch_index].updateSlot(0,i,teacher_name)
                        temp_value = value - self.no_of_days
                        self.data.iat[teacher_index, batch_index+1] = temp_value
                    
#                 print(temp_value)
#                 alloting remaining lectures
                run = self.no_of_days*(self.no_of_lectures -1)
                while (temp_value!=0 and run!=0):
                    for i in range(self.no_of_lectures):
                        if i == 0:
                            continue
                        for j in range(self.no_of_days):
                            if self.Batch_list[batch_index].isAvai(i,j) and self.Teacher_list[teacher_index].isAvai(i,j):
                                self.Teacher_list[teacher_index].updateSlot(i,j,batch_name)
                                self.Batch_list[batch_index].updateSlot(i,j,teacher_name)
                                temp_value-=1
                                self.data.iat[teacher_index, batch_index+1] = temp_value
#                                 print("I'm alloting lecture")
                                if temp_value == 0:
                                    break
                            run-=1
                        if temp_value == 0:
                                break
                
                if temp_value!=0:
                    #try to swap in between the existing lectures
                    temp_tt = self.Batch_list[batch_index].getTT()
                    for temp_row_index, temp_row in temp_tt.iterrows():
                        if temp_row_index == 0:
                            continue
                        temp_col_index = 0
                        for temp_col_name, temp_val in temp_row.items():
                            #check if the teacher is available in this slot
                            if self.Teacher_list[teacher_index].isAvai(temp_row_index, temp_col_index):
                                temp_teacher = self.searchTeacher(temp_val)
                                nan_indexes = np.where(pd.isnull(temp_tt))
#                                 print(nan_indexes)
                                # Get the indexes of all NaN values
                                nan_index_list = list(zip(nan_indexes[0], nan_indexes[1]))
#                                 print(nan_index_list)

#                                 check if temp_teacher is available for any of the vacant slot
                                for temp_index in nan_index_list:
                                    if temp_index[1] == 0:
                                        continue
#                                     print(temp_index)
#                                     print("im here")
                                    if temp_teacher.isAvai(temp_index[0], temp_index[1]):
#                                         print("swapping", temp_teacher.name, self.Teacher_list[teacher_index].name, temp_index)
                                        #updating available teacher's time table
                                        temp_teacher.updateSlot(temp_index[0], temp_index[1], batch_name)
                                        temp_teacher.updateSlot(temp_row_index, temp_col_index, np.nan)
                                        #updating current teacher's time table
                                        self.Teacher_list[teacher_index].updateSlot(temp_row_index, temp_col_index, batch_name)
                                        #updating current batch time table
                                        self.Batch_list[batch_index].updateSlot(temp_row_index, temp_col_index, self.Teacher_list[teacher_index].name)
                                        self.Batch_list[batch_index].updateSlot(temp_index[0], temp_index[1], temp_teacher.name)
                                        #updating lectures an batch timetable
                                        temp_value -= 1
                                        self.data.iat[teacher_index, batch_index+1] = temp_value
                                        temp_tt = self.Batch_list[batch_index].getTT()
                                        break
                            temp_col_index+=1
                            if temp_value == 0:
                                break
                        if temp_value == 0:
                            break
                col_index+=1
    
    def allotnanlec(self):
#         print(self.data)
        not_completed = not self.checkCompleted()
        while not_completed:
#             print('im stuck in 1st while loop')
            batch_index = -1
            for batch in self.Batch_list:
#                 print('im stuck in for loop 1')
                batch_index+=1
                status = batch.getStatus()
                if (status):
                    continue
                batch_tt = batch.getTT()
                nan_indexes = np.where(pd.isnull(batch_tt))
                nan_index_list = list(zip(nan_indexes[0], nan_indexes[1]))
                #try to fillna values
                for index, value in self.data[batch.name].iteritems():
#                     print('im stuck in for loop 1.1')
                    if value<= 0:
                        continue
                    teacher = self.Teacher_list[index]
                    for nan_index in nan_index_list:
#                         print('im stuck in for loop 1.1.1')
                        if teacher.isAvai(nan_index[0],nan_index[1]) and value>0 :
                            teacher.updateSlot(nan_index[0],nan_index[1],batch.name)
                            batch.updateSlot(nan_index[0],nan_index[1],teacher.name)
                            value-=1
                            self.data.iat[index, batch_index+1] = value
                            batch_tt = batch.getTT()
                    
                    if value<= 0:
                        continue
                    for nan_index in nan_index_list:
#                         print('im stuck in for loop 1.1.2')
                        for batch_row_index, batch_row in batch_tt.iterrows():
                            batch_col_index = -1
#                             print('im stuck in for loop 1.1.2.1')
                            for batch_col_name, teacher_name in batch_row.items():
#                                 print('im stuck in for loop 1.1.2.1.1')
                                batch_col_index+=1
#                                 print(teacher_name)
                                if pd.isna(teacher_name):
                                    continue
                                new_teacher = self.searchTeacher(teacher_name)
                                if new_teacher.isAvai(nan_index[0], nan_index[1]) and teacher.isAvai(batch_row_index, batch_col_index):
                                    #updating available teacher's time table
                                    new_teacher.updateSlot(nan_index[0], nan_index[1], batch.name)
                                    new_teacher.updateSlot(batch_row_index, batch_col_index, np.nan)
                                    #updating current teacher's time table
                                    teacher.updateSlot(batch_row_index, batch_col_index, batch.name)
                                    #updating current batch time table
                                    batch.updateSlot(batch_row_index, batch_col_index, teacher.name)
                                    batch.updateSlot(nan_index[0], nan_index[1], new_teacher.name)
                                    #updating lectures an batch timetable
                                    value -= 1
                                    self.data.iat[index, batch_index+1] = value
                                    batch_tt = batch.getTT()
                                if value<=0:
                                    break
                            if value<=0:
                                break
                        if value<=0:
                            break
                
                # if not able to fill the nan values, swap teacher between batches
                for index, value in self.data[batch.name].iteritems():
#                     print('im stuck in for loop 1.2')
                    if value<= 0:
                        continue
                    teacher = self.Teacher_list[index]
                    for nan_index in nan_index_list:
#                         print('im stuck in for loop 1.2.1')
                        new_batch_name = teacher.getVal(nan_index[0], nan_index[1])
                        new_batch, new_batch_index = self.searchBatch(new_batch_name)
                        
                        teacher.updateSlot(nan_index[0], nan_index[1], batch.name)
                        batch.updateSlot(nan_index[0], nan_index[1], teacher.name)
                        new_batch.updateSlot(nan_index[0], nan_index[1], np.nan)
                        value -= 1
                        self.data.iat[index, batch_index+1] = value
                        batch_tt = batch.getTT()
                        self.data.iat[index, new_batch_index+1]+=1
                        if value <= 0:
                            break
            not_completed = not self.checkCompleted()
#             if not not_completed:
#                 print(self.data)
        
    def showBatchwiseTT(self):
        for batch in self.Batch_list:
            print(batch.name)
            batch.showTT()
            
    def showTeacherwiseTT(self):
        for teacher in self.Teacher_list:
            print(teacher.name)
            teacher.showTT()
            
    def searchTeacher(self, name):
        for teacher in self.Teacher_list:
            if teacher.name == name:
                return teacher
            
    def searchBatch(self, name):
        index = -1
        for batch in self.Batch_list:
            index+=1
            if batch.name == name:
                return batch,index
```

```python
import time

time_table = TimeTable(data, batches, No_of_working_days, No_of_lectures_per_day) #no of days, no of lectures per day

start_time = time.time()

time_table.createAllFiles()
time_table.allocateLectures()
time_table.allotnanlec()
end_time = time.time()

execution_time = end_time - start_time
print(f"Execution time: {execution_time} seconds")
```
