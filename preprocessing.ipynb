```python
import pandas as pd
import numpy as np

# Load the DataFrame from a CSV file
data = pd.read_csv('constraints4.csv') # test file name in the same directory 
df = pd.DataFrame(data)

# Calculate the number of non-zero values in each column and row
non_zero_columns = df.apply(lambda col: col.astype(bool).sum())
non_zero_rows = df.apply(lambda row: row.astype(bool).sum(), axis=1)

# Sort columns based on the number of non-zero values
sorted_columns = non_zero_columns.sort_values(ascending=False).index

# Sort rows based on the number of non-zero values and sum of columns in case of tie
sorted_rows = non_zero_rows.sort_values(ascending=False, kind='mergesort').index

# Create a new DataFrame with sorted columns and rows
sorted_df = df.loc[sorted_rows, sorted_columns]

print("Original DataFrame:")
print(df)
print("\nSorted DataFrame with columns and rows based on non-zero values count and sum of all column values:")
print(sorted_df)

# Save the sorted DataFrame to a new CSV file
file_name = "sorted_constraints.csv"
sorted_df.to_csv(file_name, index=False)
```
